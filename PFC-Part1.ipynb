```python
elementos=10
print(pow(elementos,5))
```

```python
wipeAnalysis()
paso=(0.9)/(elementos)
precision=8

ff = open ('datasetICC.txt','w')
ff2 = open ('respICC.txt','w')

ff3 = open ('datasetICC-ABS.txt','w')

ff4 = open ('datasetICC-SRSS.txt','w')

ff5 = open ('datasetICC-CQC.txt','w')


nro=0

for i in range(elementos):
    for j in range(elementos):
        for k in range(elementos):
            for l in range(elementos):
                for m in range(elementos):
                    cadena=""
                    cadena2=""
                    cadena3=""
                    cadena4=""
                    cadena5=""
                    
                    nro=nro+1
                    print("corrida numero "+str(nro))
                    
                    dam1=paso*(i+1)
                    dam2=paso*(j+1)
                    dam3=paso*(k+1)
                    dam4=paso*(l+1)
                    dam5=paso*(m+1)
                    
                    cadena2=cadena2+str(round(dam1,precision))+" "
                    cadena2=cadena2+str(round(dam2,precision))+" "
                    cadena2=cadena2+str(round(dam3,precision))+" "
                    cadena2=cadena2+str(round(dam4,precision))+" "
                    cadena2=cadena2+str(round(dam5,precision))+"\n"
                    
                    
                    Ac=63.41
                    Ic=320.0
                    Ab=63.41
                    Ib=10e+12 
                    E=30000

                    #Primer Piso
                    element('elasticBeamColumn',1,1,3,Ac*dam1*dam1,E,2*Ic*dam1*dam1*dam1,1)
                    element('elasticBeamColumn',6,2,4,Ac*dam1*dam1,E,2*Ic*dam1*dam1*dam1,1)

                    #Segundo Piso
                    element('elasticBeamColumn',2,3,5,Ac*dam2*dam2,E,2*Ic*dam2*dam2*dam2,1)
                    element('elasticBeamColumn',7,4,6,Ac*dam2*dam2,E,2*Ic*dam2*dam2*dam2,1)

                    #Tercer Piso
                    element('elasticBeamColumn',3,5,7,Ac*dam3*dam3,E,2*Ic*dam3*dam3*dam3,1)
                    element('elasticBeamColumn',8,6,8,Ac*dam3*dam3,E,2*Ic*dam3*dam3*dam3,1)

                    #Cuarto Piso
                    element('elasticBeamColumn',4,7,9,Ac*dam4*dam4,E,2*Ic*dam4*dam4*dam4,1)
                    element('elasticBeamColumn',9,8,10,Ac*dam4*dam4,E,2*Ic*dam4*dam4*dam4,1)

                    #Quinto Piso
                    element('elasticBeamColumn',5,9,11,Ac*dam5*dam5,E,2*Ic*dam5*dam5*dam5,1)
                    element('elasticBeamColumn',10,10,12,Ac*dam5*dam5,E,2*Ic*dam5*dam5*dam5,1)

                    #Vigas del Modelo
                    element('elasticBeamColumn',11,3,4,Ac,E,Ib,1)
                    element('elasticBeamColumn',12,5,6,Ac,E,Ib,1)
                    element('elasticBeamColumn',13,7,8,Ac,E,Ib,1)
                    element('elasticBeamColumn',14,9,10,Ac,E,Ib,1)
                    element('elasticBeamColumn',15,11,12,Ac,E,Ib,1)

                    #recorder('Node', '-file', "vibracion/modes.out", '-closeOnWrite', '-nodeRange',1,12, '-dof',1, 'eigen 1')

                    pi=3.141593
                    
                    lam=eigen(numModes)

                    T1=2*pi/math.sqrt(lam[0])
                    T2=2*pi/math.sqrt(lam[1])                    
                    
                    
                    cadena=cadena+str(round(T1,precision))+" "
                    cadena3=cadena3+str(round(abs(T1)+abs(T2),precision))+" "
                    cadena4=cadena4+str(round(math.sqrt(pow(T1,2)+pow(T2,2)),precision))+" "
                    cadena5=cadena5+str(round(0.25*(abs(T1)+abs(T2))+0.75*(math.sqrt(pow(T1,2)+pow(T2,2))),precision))+" "                  
                    
                    cadena=cadena+str(round(nodeEigenvector(1,1,1),precision))+" "
                    cadena=cadena+str(round(nodeEigenvector(3,1,1),precision))+" "
                    cadena=cadena+str(round(nodeEigenvector(5,1,1),precision))+" "
                    cadena=cadena+str(round(nodeEigenvector(7,1,1),precision))+" "
                    cadena=cadena+str(round(nodeEigenvector(9,1,1),precision))+"\n"
                    
                    cadena3=cadena3+str(round(   abs(nodeEigenvector(1,1,1))+abs(nodeEigenvector(1,2,1))  ,precision))+" "
                    cadena3=cadena3+str(round(   abs(nodeEigenvector(3,1,1))+abs(nodeEigenvector(3,2,1))  ,precision))+" "
                    cadena3=cadena3+str(round(   abs(nodeEigenvector(5,1,1))+abs(nodeEigenvector(5,2,1))  ,precision))+" "
                    cadena3=cadena3+str(round(   abs(nodeEigenvector(7,1,1))+abs(nodeEigenvector(7,2,1))  ,precision))+" "
                    cadena3=cadena3+str(round(   abs(nodeEigenvector(9,1,1))+abs(nodeEigenvector(9,2,1))  ,precision))+"\n"
                    
                    cadena4=cadena4+str(round(   math.sqrt(pow(nodeEigenvector(1,1,1),2)+pow(nodeEigenvector(1,2,1),2))  ,precision))+" "
                    cadena4=cadena4+str(round(   math.sqrt(pow(nodeEigenvector(3,1,1),2)+pow(nodeEigenvector(3,2,1),2))  ,precision))+" "
                    cadena4=cadena4+str(round(   math.sqrt(pow(nodeEigenvector(5,1,1),2)+pow(nodeEigenvector(5,2,1),2))  ,precision))+" "
                    cadena4=cadena4+str(round(   math.sqrt(pow(nodeEigenvector(7,1,1),2)+pow(nodeEigenvector(7,2,1),2))  ,precision))+" "
                    cadena4=cadena4+str(round(   math.sqrt(pow(nodeEigenvector(9,1,1),2)+pow(nodeEigenvector(9,2,1),2))  ,precision))+"\n"
                    
                    cadena5=cadena5+str(round(  0.25*(abs(nodeEigenvector(1,1,1))+abs(nodeEigenvector(1,2,1)))+0.75*(math.sqrt(pow(nodeEigenvector(1,1,1),2)+pow(nodeEigenvector(1,2,1),2)))   ,precision))+" "
                    cadena5=cadena5+str(round(   0.25*(abs(nodeEigenvector(3,1,1))+abs(nodeEigenvector(3,2,1)))+0.75*(math.sqrt(pow(nodeEigenvector(1,1,1),2)+pow(nodeEigenvector(1,2,1),2)))  ,precision))+" "
                    cadena5=cadena5+str(round(   0.25*(abs(nodeEigenvector(5,1,1))+abs(nodeEigenvector(5,2,1)))+0.75*(math.sqrt(pow(nodeEigenvector(1,1,1),2)+pow(nodeEigenvector(1,2,1),2)))  ,precision))+" "
                    cadena5=cadena5+str(round(   0.25*(abs(nodeEigenvector(7,1,1))+abs(nodeEigenvector(7,2,1)))+0.75*(math.sqrt(pow(nodeEigenvector(1,1,1),2)+pow(nodeEigenvector(1,2,1),2)))  ,precision))+" "
                    cadena5=cadena5+str(round(   0.25*(abs(nodeEigenvector(9,1,1))+abs(nodeEigenvector(9,2,1)))+0.75*(math.sqrt(pow(nodeEigenvector(1,1,1),2)+pow(nodeEigenvector(1,2,1),2)))  ,precision))+"\n"
                    
                    integrator("LoadControl", 0, 1,0,0)
                    test("EnergyIncr",1.0e-10,100,0)
                    algorithm("Newton")
                    numberer("RCM")
                    constraints("Transformation")
                    system("ProfileSPD")
                    analysis("Static")
                    res=analyze(1)
                    
                    ff.write(cadena)
                    ff2.write(cadena2)
                    ff3.write(cadena3)
                    ff4.write(cadena4)                    
                    ff5.write(cadena5)
                    
                    if(res<0):
                        print("Eror Modal Analysis")

                    remove('ele',1)
                    remove('ele',2)
                    remove('ele',3)
                    remove('ele',4)
                    remove('ele',5)
                    remove('ele',6)
                    remove('ele',7)
                    remove('ele',8)
                    remove('ele',9)
                    remove('ele',10)
                    remove('ele',11)
                    remove('ele',12)
                    remove('ele',13)
                    remove('ele',14)
                    remove('ele',15)

                    wipeAnalysis()

ff.close()
ff2.close()
print("Termino")
```
